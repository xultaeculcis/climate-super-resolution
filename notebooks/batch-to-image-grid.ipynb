```python
import os
from glob import glob

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import rasterio as rio
import torch
from matplotlib import colors

%matplotlib inline
```

```python
batch = torch.rand(64, 1, 128, 128)
```

```python
def make_grid(I, ncols=8):
    nimg = I.shape[0]
    H = I.shape[2]
    W = I.shape[3]
    ncols = min(nimg, ncols)
    nrows = int(np.ceil(float(nimg) / ncols))
    canvas = np.zeros((H * nrows, W * ncols), dtype=I.dtype)
    i = 0
    for y in range(nrows):
        for x in range(ncols):
            if i >= nimg:
                break
            canvas[y * H : (y + 1) * H, x * W : (x + 1) * W] = I[i]
            i = i + 1
    return canvas


np_grid_img = make_grid(batch.cpu().numpy())
plt.imshow(np_grid_img, cmap="jet")
```
